```python
import pandas as pd
file = 'data/gender_age_train.csv' 
gender = pd.read_csv(file)
gender.head(2)
```

```python
file = 'data/phone_brand_device_model.csv' 
phones = pd.read_csv(file).drop_duplicates()
phones = phones.drop(phones[phones.device_id.duplicated()].index.tolist())
phones.head(2)
```

```python
gender_phone=pd.merge(gender,phones,how="inner",on="device_id")
gender_phone.head(2)
```

```python
X_values=gender_phone[['phone_brand','device_model']]
y_values=gender_phone['group']
```

```python
X_ohc =pd.get_dummies(X_values)
```

```python
y_ohc=pd.get_dummies(y_values)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_ohc, y_values, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred_class_logreg = logreg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred_class_logreg)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred_class_gnb = gnb.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_class_gnb)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train) 
```

```python
y_pred_knn = neigh.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred_knn)
```
