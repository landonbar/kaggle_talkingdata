```python
import pandas as pd
file = 'data/gender_age_train.csv' 
gender = pd.read_csv(file)
gender.head(2)
```

```python
file = 'data/phone_brand_device_model.csv' 
phones = pd.read_csv(file).drop_duplicates()
phones = phones.drop(phones[phones.device_id.duplicated()].index.tolist())
phones.head(2)
```

```python
ohc = pd.get_dummies(phones[['phone_brand','device_model']])
```

```python
ohc.loc[:,'device_id'] = phones.device_id
```

```python
gender_phone=pd.merge(gender,ohc,how="inner",on="device_id")
gender_phone.head(2)
```

```python
y_values=gender_phone['group']
```

```python
y_ohc=pd.get_dummies(y_values)
```

```python
X_values=gender_phone.drop(['device_id', 'gender','age','group'], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_values, y_values)
```

```python
file = 'data/gender_age_test.csv' 
sub = pd.read_csv(file)
sub.head(2)
```

```python
sub_phone=pd.merge(sub,ohc,how="inner",on="device_id")
sub_phone.head(2)
```

```python
X_test=sub_phone.drop(['device_id'], axis=1)
```

```python
y_pred_class_logreg_prob = logreg.predict_proba(X_test)
```

```python
pred=pd.DataFrame(y_pred_class_logreg_prob, columns=logreg.classes_)
```
